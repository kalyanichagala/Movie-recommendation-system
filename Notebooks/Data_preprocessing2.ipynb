# 2017 Movies


```python
#Extracting only 2017 data as there is less data available in 2018,2019,2020 & 2021
movie17=moviedata2[moviedata2["release_date"]==2017.0]
```

```python
movie17.head()
```

```python
#Reading movie credits data 
credits=pd.read_csv("creditsmovie.csv")
```

```python
credits.head()
```

```python
import ast
```

```python
credits["crew"]=credits["crew"].apply(lambda x : ast.literal_eval(x))
credits["cast"]=credits["cast"].apply(lambda x : ast.literal_eval(x))
movie17["id"]=movie17["id"].apply(lambda x : ast.literal_eval(x))
```

```python
final=pd.merge(credits,movie17,on="id",how="inner")
```

```python
pd.set_option("display.max_columns",None)
```

```python
final.head()
```

```python
final.shape
```

```python
director=[]
for x in final["crew"]:
    mix=[]
    for f in x:
        for h in f:
            if f[h] == "Director":
                mix.append(f["name"])
    director.append(mix)
            
```

```python
actorname1=[]
for x in final["cast"]:
    if x == []:
        y=np.nan
    else:
        y=x[0]["name"]
    actorname1.append(y)
```

```python
actorname2=[]
for h in final["cast"]:
    if len(h)<2:
        y=np.nan
    else:
        y=h[1]["name"]
    actorname2.append(y)
        
```

```python
actorname3=[]
for h in final["cast"]:
    if len(h)<3:
        y=np.nan
    else:
        y=h[2]["name"]
    actorname3.append(y)
```

```python
final["genres"]=final["genres"].apply(lambda x : ast.literal_eval(x))
```

```python
genre=[]
for t in final["genres"]:
    grp=[]
    for l in t :
        for d in l :
            if d == "name":
                grp.append(l[d])
    genre.append(grp)
            
            
        
        
    
```

```python
final["genre"]=genre
```

```python
final["genre"]=final["genre"].apply(lambda x : " ".join(x))
```

```python
final["director_name"]=director
```

```python
final["actor_1_name"]=actorname1
```

```python
final["actor_2_name"]=actorname2
```

```python
final["actor_3_name"]=actorname3
```

```python
finalnew2=pd.DataFrame({})
```

```python
finalnew2[["director_name","actor_1_name","actor_2_name","actor_3_name","movie_title","genres"]]=final[["director_name","actor_1_name","actor_2_name","actor_3_name","title","genre"]]
```
